```python
import re
def clean_text(text):
    text = re.sub(r"@[A-Za-z0-9]+", "", text)
    text = re.sub(r"[^\x00-\x7F]+", "", text)
    text = text.lower()
    return text
```

```python
dataset['clean text '] = dataset.tweet.apply(lambda x: clean_text(x))
```

```python
def gen_freq(text):
    word_list = []
    # Define 'stop' within the function's scope:
    stop = stopwords.words('english')
    for tw_word in text.split():
        word_list.extend(tw_word)
        word_freq = pd.Series(word_list).value_counts()
        word_freq = word_freq.drop(stop, errors='ignore')
    return word_freq
```

```python
def any_neg(words):
    for word in words:
        if word in ['n', 'no', 'non', 'not'] or re.search(r"\wn't", word):
            return 1
    else:
        return 0
```

```python
def any_rare(words, rare_100):
    for word in words:
        if word in rare_100:
            return 1
    else:
        return 0
```

```python
def is_question(words):
    for word in words:
        if word in ['when', 'what', 'how', 'why', 'who']:
            return 1
    else:
        return 0
```
